```python
def html(func):
    def wrapper(*args, **kwargs):
        print("<html>")
        func(*args, **kwargs)
        print("</html>")
    return wrapper
 
def body(func):
    def wrapper(*args, **kwargs):
        print("<body>")
        func(*args, **kwargs)
        print("</body>")
    return wrapper
 
@html
@body
def text_printer(text):
    print(text)
 
text_printer("This is my text")
```

`<html> <body> This is my text </body> </html`

```python
def divide_numbers(a, b):
    try:
        result = a / b
    except ZeroDivisionError as e:
        raise ValueError("Cannot divide by zero") from e
    return result
try:
    result = divide_numbers(5, 0)
    print("Result:", result)
except ValueError as e:
    print("Error:", e)
    print("Original exception:", e.__cause__)
```

Error: Cannot divide by zero 
Original Exception: division by zero
