## Pep Quiz 

### 1) Imports 
- avoid wildcards 
- imports on sep lines 

---

### 2) PEP 8 Blank lines 
- __two__ blank lines for toplevel function and class 
- __one__ blank line for methods inside class 

---

### 3) PEP?
Python Enhancement Proposal that's part of the Python Dev Guide doc (all) 

---

### 4) Zen of Pyton 
```python
import this
```

---

### 5) Zen Philosophy with Namespace 

Generally **AVOID** `global` keyword to modify var inside func because of name collisions 

---

### 6) Zen Philosophy with Exception Handling 

Use `except ImportError as e` and `print(e)` for exception info 

--- 

### 7) Py Docstrings regarding PEP 257 
- Docstrings occur first statement of func/class/method/module 
- Multiline doc should have summary line first, blank then elaborate 

---

### 8) PEP 8 Guidelines 
- Follow PEP 8 guidelines if starting new Py Project 
- Ignore PEP8 guidelines if break backwards compatability or degrade code readability 

---

### 9) PEP 8 Rec for comments, inline, blocks?
- Comments are complete sentences (cap and period) 
- Block comments refer to the code that follows them 

---

### 10) `None` objects with PEP 8

Use `is` instead of `==`

---

### 11) PEP that describe new lang features?

Standard Track PEP 

---

### 12) Naming Styles for constants?

Upper-case with underscores (`MY_SAMPLE_CONSTANT`)

--- 

### 13) "Flat is better than Nested"

Nested makes code harder to follow (try to not have more than **three** levels

--- 

### 14) PEP 8 Order of imports 

Standard, Third-party, Local
